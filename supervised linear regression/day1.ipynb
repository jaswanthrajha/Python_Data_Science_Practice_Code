```python
y = bos.Price
x = bosl
xc = sm.add_constant(x)
model = sm.OLS(y,xc).fit()
model.summary()
```

```python
cols= list(xc.columns)
while (len(cols)>0):
    x1=xc[cols]
    model = sm.OLS(y,x1).fit()
    p = model.pvalues
    max_p = max(p)
    feature_maxp = p.idxmax()
    
    if max_p >0.5:
        cols.remove(feature_maxp)
        print('feature removed',feature_maxp,'prob: ',max_p)
    else:
        break
```
