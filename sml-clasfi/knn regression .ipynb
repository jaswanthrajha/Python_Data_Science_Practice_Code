```python
x = df.drop(['mpg','disp','acc', 'yr'],axis=1)
y = df.mpg
```

```python
x
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30, random_state=2)
```

```python
x_train
```

```python
x_test
```

```python
y_test
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsRegressor

lr = KNeighborsRegressor(n_neighbors=29, weights='uniform',metric='manhattan')
lr.fit(x_train,y_train)
```

```python
y_pred_train= lr.predict(x_train)
```

```python
y_pred_test= lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y_train,y_pred_train)*100
```

```python
r2_score(y_test,y_pred_test)*100
```

```python
np.sqrt(mean_squared_error(y_train,y_pred_train))
```

```python
np.sqrt(mean_squared_error(y_test,y_pred_test))
```

```python
from sklearn.model_selection import GridSearchCV
params = {'n_neighbors': np.arange(1,50),'metric' : ['euclidean','manhattan']}
knn = KNeighborsRegressor()
gs = GridSearchCV(estimator=knn,param_grid=params,cv=5)
gs.fit(x_train,y_train)
print(gs.best_params_)
```
