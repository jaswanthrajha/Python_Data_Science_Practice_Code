```python
for i in a.select_dtypes(exclude=object):
    print(i)
    
    print('\nmean   : ',(a[i].mean()))
    print('median : ',a[i].median())
    print('mode   : ',a[i].mode()[0])
    print('sd     : ',a[i].std())
    print('skew   : ',a[i].skew())
    print('kurt   : ',a[i].kurt())
    print('range  : ',a[i].max()-a[i].min())
    print('\ndescribe\n')
    print(a[i].describe())
    sns.boxplot(a[i])
    plt.show()
    
    
    print('----------------------------------------------------------------------------------------------------------------')
    
```

```python
for i in a.select_dtypes(exclude=object):
    print(i)
    print('q1  : ',a[i].quantile(.25))
    print('q3  : ',a[i].quantile(.75))
    print('iqr : ',a[i].quantile(.75)-a[i].quantile(.25))
    
    print('lower wisker',a[i].quantile(.25)-1.5*(a[i].quantile(.75)-a[i].quantile(.25)))
    
    l = a[i].quantile(.25)-1.5*(a[i].quantile(.75)-a[i].quantile(.25))
    
    print('upper wisker',a[i].quantile(.75)+1.5*(a[i].quantile(.75)-a[i].quantile(.25)))
    
    u = a[i].quantile(.75)+1.5*(a[i].quantile(.75)-a[i].quantile(.25))
    
    if (a[(a[i]<l )|( a[i]>u)].shape[0]!=0):
        print(a[(a[i]>l) & (a[i]<u)])

    else:
        print(a[i])
        
    
    
```
